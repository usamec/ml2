```python
import torch
import matplotlib.pyplot as plt
```

```python
# This is how we initialized neural networks before ~2010. We could not train deeper than 3 layers network
# at the time.

def run_network(h=128):
    x = torch.randn(1024,h) 
    # Everybody would use nn.linear here, but this is 2008
    w1 = torch.randn(h,h) * 0.1
    w2 = torch.randn(h,h) * 0.1
    w3 = torch.randn(h,h) * 0.1
    
    h1 = torch.relu(x.matmul(w1))
    h2 = torch.relu(h1.matmul(w2))
    h3 = torch.relu(h2.matmul(w3))
    return h1, h2, h3

h1, h2, h3 = run_network(h=32)
```

```python
# Observe that for small network each layer gets smaller and smaller features
# (vanishing)

plt.plot(h1.std(dim=0), label="h1")
plt.plot(h2.std(dim=0), label="h2")
plt.plot(h3.std(dim=0), label="h3")
plt.yscale("log")
plt.legend()
```

```python
# But for large network features get larger and larget (they explode)
# This is not good and makes network hard to train

h1, h2, h3 = run_network(h=1024)
plt.plot(h1.std(dim=0)[:100], label="h1")
plt.plot(h2.std(dim=0)[:100], label="h2")
plt.plot(h3.std(dim=0)[:100], label="h3")
plt.yscale("log")
plt.legend()
```
